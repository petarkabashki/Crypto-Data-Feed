## Convert json to Parquet 

```python
import os
import pandas as pd

# Define source and destination directories
source_dir = 'freq-user-data/data'
destination_dir = 'data-parquet'

# Function to convert JSON to Parquet maintaining the directory structure
def convert_json_to_parquet(source_dir, destination_dir):
    for root, dirs, files in os.walk(source_dir):
        for file in files:
            if file.endswith('.json'):
                # Construct full file path
                file_path = os.path.join(root, file)
                
                # Read the JSON file into a DataFrame
                df = pd.read_json(file_path)
                
                # Determine the relative path and create corresponding destination directory
                relative_path = os.path.relpath(root, source_dir)
                destination_subdir = os.path.join(destination_dir, relative_path)
                os.makedirs(destination_subdir, exist_ok=True)
                
                # Construct the output Parquet file path
                parquet_filename = f"{os.path.splitext(file)[0]}.parquet"
                parquet_file_path = os.path.join(destination_subdir, parquet_filename)
                
                # Save the DataFrame to a Parquet file
                df.to_parquet(parquet_file_path, engine='pyarrow', index=False)
                
                print(f"Converted {file_path} to {parquet_file_path}")

print("Starting conversion...")
convert_json_to_parquet(source_dir, destination_dir)
print("All files have been successfully converted to Parquet format.")
```
